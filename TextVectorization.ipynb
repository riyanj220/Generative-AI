## **TF-IDF**

```python
df = pd.DataFrame({"text":["my name is riyan",
                         "people watch riyan",
                         "hello guys welcome me",
                          "welcome riyan welcome"],"output":[1,1,0,0]})

df
```

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
```

```python
def remove_stopwords(text):
    new_text = []

    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    x = new_text[:]
    new_text.clear()
    return " ".join(x)
```

```python
df['text'] = df['text'].apply(remove_stopwords)
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfid = TfidfVectorizer()
```

```python
arr = tfid.fit_transform(df['text']).toarray()
```

```python
arr
```

```python
print(tfid.idf_)
```

```python
import os
from nltk.tokenize import sent_tokenize
from gensim.utils import simple_preprocess

story = []
folder_path = '/content/drive/MyDrive/GenerativeAINotebooks/datasets/game_of_thrones'

for filename in os.listdir(folder_path):
    if filename.endswith('.txt'):
        file_path = os.path.join(folder_path, filename)
        try:
            with open(file_path, 'r', encoding='utf-8') as f:
                corpus = f.read()
        except UnicodeDecodeError:
            with open(file_path, 'r', encoding='latin-1') as f:
                corpus = f.read()
        raw_sent = sent_tokenize(corpus)
        for sent in raw_sent:
            story.append(simple_preprocess(sent))

print(f"Total sentences collected: {len(story)}")
```

```python

story
```

```python
len(story)
```

```python
story[0]
```

```python
model = gensim.models.Word2Vec(
    window=10
    min_count=2 # words that appear less than 2 times in my documents will be ignored
)
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count , epochs=model.epochs)
# total+examples Number of sentences to train on (auto-calculated).
# epochs = number of times to go through data
```

```python
model.wv.most_similar('daenerys')
```

### Performing basic pre processing
    - Remove tags-HTML
    - Lower case
    - stop words removing

```python
df['Review'] = df['Review'].str.lower()
```

```python
df['Review'][3]
```

```python
import re
def remove_html_tags(text):
  pattern = re.compile('<.*?>')
  return pattern.sub(r'', text)
```

```python
df['Review'] = df['Review'].apply(remove_html_tags)
```

```python
def remove_stopwords(text):
    new_text = []

    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    x = new_text[:]
    new_text.clear()
    return " ".join(x)
```

```python
df['Review'] = df['Review'].apply(remove_stopwords)
```

```python
df['Review'][0]
```

```python
X = df.iloc[:,0:1]
y = df['Sentiment']
```

```python
X.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y =encoder.fit_transform(y)
```

```python
y
```

```python
X_train,X_test,Y_train, y_test = train_test_split(X,y , test_size=0.2, random_state=1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# Applying Bag of words
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
X_train_bow = cv.fit_transform(X_train['Review']).toarray()   # fit_tranform is used when we apply count vectorizer on training data
X_test_bow = cv.transform(X_test['Review']).toarray()         # transform is used when we apply count vectorizer on testing data
```

```python
X_train_bow
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train_bow, Y_train)
```

```python
y_pred = gnb.predict(X_test_bow)

from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()

rf.fit(X_train_bow,Y_train)

y_pred = rf.predict(X_test_bow)

accuracy_score(y_test,y_pred)
```

```python
cv = CountVectorizer(max_features=3000)  # max_features means we are taking 3000 frequent unique words

X_train_bow = cv.fit_transform(X_train['Review']).toarray()
X_test_bow = cv.transform(X_test['Review']).toarray()

rf = RandomForestClassifier()

rf.fit(X_train_bow,Y_train)
y_pred = rf.predict(X_test_bow)
accuracy_score(y_test,y_pred)
```

```python
# USING N GRAMS
cv = CountVectorizer(ngram_range=(1,2),max_features=5000)

X_train_bow = cv.fit_transform(X_train['Review']).toarray()
X_test_bow = cv.transform(X_test['Review']).toarray()

rf = RandomForestClassifier()

rf.fit(X_train_bow,Y_train)
y_pred = rf.predict(X_test_bow)
accuracy_score(y_test,y_pred)
```

```python
# USING TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
```

```python
X_train_tfidf = tfidf.fit_transform(X_train['Review']).toarray()
X_test_tfidf = tfidf.transform(X_test['Review'])
```

```python
rf = RandomForestClassifier()
rf.fit(X_train_tfidf,Y_train)
y_pred = rf.predict(X_test_tfidf)
accuracy_score(y_test,y_pred)
```
