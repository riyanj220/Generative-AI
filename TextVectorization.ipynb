## **TF-IDF**

```python
df = pd.DataFrame({"text":["my name is riyan",
                         "people watch riyan",
                         "hello guys welcome me",
                          "welcome riyan welcome"],"output":[1,1,0,0]})

df
```

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
```

```python
def remove_stopwords(text):
    new_text = []

    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    x = new_text[:]
    new_text.clear()
    return " ".join(x)
```

```python
df['text'] = df['text'].apply(remove_stopwords)
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfid = TfidfVectorizer()
```

```python
arr = tfid.fit_transform(df['text']).toarray()
```

```python
arr
```

```python
print(tfid.idf_)
```

```python
import os
from nltk.tokenize import sent_tokenize
from gensim.utils import simple_preprocess

story = []
folder_path = '/content/drive/MyDrive/GenerativeAINotebooks/datasets/game_of_thrones'

for filename in os.listdir(folder_path):
    if filename.endswith('.txt'):
        file_path = os.path.join(folder_path, filename)
        try:
            with open(file_path, 'r', encoding='utf-8') as f:
                corpus = f.read()
        except UnicodeDecodeError:
            with open(file_path, 'r', encoding='latin-1') as f:
                corpus = f.read()
        raw_sent = sent_tokenize(corpus)
        for sent in raw_sent:
            story.append(simple_preprocess(sent))

print(f"Total sentences collected: {len(story)}")
```

```python
len(story)
```

```python
story[0]
```

```python
model = gensim.models.Word2Vec(
    window=10
    min_count=2 # words that appear less than 2 times in my documents will be ignored
)
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count , epochs=model.epochs)
# total+examples Number of “sentences” to train on (auto-calculated).
# epochs = number of times to go through data
```

```python
model.wv.most_similar('daenerys')
```
